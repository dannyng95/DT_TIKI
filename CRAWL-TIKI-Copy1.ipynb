```python
#Build a function
# Import Library

import pandas as pd
import requests
import random
import time
from time import sleep
from bs4 import BeautifulSoup
from random import randint


def get_url(url):
    """Get parsed HTML from url
      Input: url to the webpage
      Output: Parsed HTML text of the webpage
    """
    # Send GET request
    r = requests.get(url)

    # Parse HTML text
    soup = BeautifulSoup(r.text, 'html.parser')

    return soup
def crawling_product(urlxx,number_of_pages,pdn):
    """Scrape the page of product you want to
      Input: url to the webpage. Default: https://tiki.vn/may-anh/c1801?src=c.1801.hamburger_menu_fly_out_banner&page=1
      Output: A list containing scraped data of all product & price
    """

    # Get parsed HTML
    dataall = []
    
    for x in range (1,number_of_pages):
        url=f"{urlxx}={x}"
        soup = get_url(url)
        
        # Find all products in the page
        products = soup.find_all('div',{'class':'product-item'})
        #Find the final page:
        if list(products) == None:
            break
        else:
        
        # List containing data of all articles
        data = []
        for product in products:
            d = {'Product-id':'','Seller-id':'','Product-SKU':'','Product-Title':'',
                 'Product-Brand':'','URL':'','img_URL':'','Regular-Price':'','Final-Price':'',
                 'TIKI-NOW':'','Installment':'','Page':''}
            try:
                d['Product-id'] = product.a['data-id']
                d['Product-SKU'] = product['product-sku']
                d['Seller-id'] = product['data-seller-product-id']
                d['Product-Title'] = product.a['title']
                d['Product-Brand'] = product['data-brand']
                d['URL'] = product.a['href']
                d['img_URL'] = product.a.div.span.img['src']
                Regularprice = product.find('span',{'class':'price-regular'})
                d['Regular-Price'] = Regularprice.text
                Finalprice = product.find('span',{'class':'final-price'})
                d['Final-Price'] = Finalprice.text.strip(" \n, ...")
                d['Page'] = x
                try:
                    Installment = product.find('span',{'class':'installment-price-v2'})
                    d['Installment'] = Installment.text
                except:
                    d['Installment'] = None
                    
                    
                try:
                    tikinow = product.find('div',{'class':'badge-service'})
                    badge_check = tikinow.div.img['src']
                    if str(badge_check) == "https://salt.tikicdn.com/ts/upload/9f/32/dd/8a8d39d4453399569dfb3e80fe01de75.png":
                        d['TIKI-NOW'] = "YES"
                except:
                    d['TIKI-NOW'] = "NO"
                data.append(d)
            except:
                print("We got one article error!")
        
        dataall.extend(data)
        sleeptime = random.uniform(1,2)
        sleep(sleeptime)

#         #products = pd.DataFrame(data = data, columns = data[0].keys())
        
        #products.to_csv(f"./Tiki_{pdn}{x}.csv", index=False)  
        full_products = pd.DataFrame(data = dataall, columns = data[0].keys())
        full_products.to_csv(f'./Full_Products_{pdn}.csv', index=False)
    return full_products
```
